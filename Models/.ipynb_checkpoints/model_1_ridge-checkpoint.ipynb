```python
import pandas as pd
import numpy as np
import pickle 
import os
import matplotlib
%matplotlib inline
# import statsmodels.api as sm
# import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import RidgeCV

from collections import OrderedDict
```

* **include all numeric and dummy features**
* **add sqrt, quadratic and cubic of numerical features**
* y as it is (linear)

```python
def power_sqrt(x):
    return x**0.5

def power2(x):
    return x**2

def power3(x):
    return x**3


def createFeatures(df):
    
    y = df['trip_count']
    X = df.drop(['trip_count', 'date_time'],1)
    
    X_num = X.iloc[:,0:8]
    X_dummy = X.iloc[:,9:]
    
    column_name = X_num.columns
    
    # create sqrt features
    for i in range(0,8):
        col_name = column_name[i]+'_sqrt'
        X_num[col_name] = X.iloc[:,i].apply(power_sqrt)
        
    # create square features
    for i in range(0,8):
        col_name = column_name[i]+'_pow2'
        X_num[col_name] = X.iloc[:,i].apply(power2)
        
    # create square features
    for i in range(0,8):
        col_name = column_name[i]+'_pow3'
        X_num[col_name] = X.iloc[:,i].apply(power3)
    
    X = pd.concat([X_num, X_dummy], axis=1)
    
    return X, y

def createTrainTest(X, y):
    X_train, X_test, y_train, y_test \
        = train_test_split(X, y, test_size=0.3, random_state=42)
    return X_train, X_test, y_train, y_test 

def performRidgeCV(X, y):
    rcv = RidgeCV(cv=10, alphas=[0.5, 0.1, 0.05, 0.01, 0.005, 0.001])
    model = rcv.fit(X, y)
    Rsq= model.score(X, y)
    
    alpha_ = model.alpha_
    return Rsq, alpha_
    
```

```python
result_dict = OrderedDict()
dict_keys = ['january', 'february', 'march', 'april', 
            'may', 'june', 'july', 'august',
            'september', 'october', 'november', 'december']
for key in dict_keys:
    result_dict[key] = []


feature_path = '../Features/Data/'
### loop through the file names
file_names = os.listdir(feature_path)

for filename in file_names:
    month_name = filename.split(".")[0]
    
    with open(feature_path+filename, 'rb') as picklefile: 
        df = pickle.load(picklefile)
        df = df.dropna()
        
    
    X, y = createFeatures(df)
    
    Rsq, alpha = performRidgeCV(X, y)
    result_dict[month_name] = [Rsq, alpha]
```

```python
pd.set_option('precision', 3)
result_df = pd.DataFrame(result_dict, index =['r_sq', 'alpha'])
result_df
```

```python
feature_path = '../Data/'
with open(feature_path+'june.p', 'rb') as picklefile: 
    df = pickle.load(picklefile)
    df = df.dropna()
        
X, y = createFeatures(df)
    
rcv = RidgeCV(cv=10, alphas=[0.5])
model = rcv.fit(X, y)
y_hat = rcv.fit(X, y).predict(X)
y_hat.shape, y.shape

from matplotlib import pyplot as plt
% matplotlib inline
fig, ax = plt.subplots(figsize=(10,5))
plt.scatter(y, y_hat)
plt.xlabel('Observed', fontsize=16)
plt.ylabel('Predicted', fontsize=16)
plt.title('Number of Rides', fontsize=18)
plt.legend(['June'], loc='upper left', fontsize=18)

plt.show()
```
